```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('/content/OUSUS 02.csv')

# Load the Excel file into a pandas DataFrame
excel_df = pd.read_excel('/content/FHEMIG_OUVIDORIAS.xlsx')

# Get the list of values from the specified column in the Excel file
ouvidoria_list = excel_df['NOME_OUVIDORIA_DESTINO'].tolist()

# Filter the DataFrame using the list of values from the Excel file
filtered_df = df[
    (df['NOME_OUVIDORIA_DESTINO'].isin(ouvidoria_list)) &
    (df['TIPIFICAÇÃO'].str.contains('TRABALHADOR EM SA', na=False))
]

# Display the filtered DataFrame
display(filtered_df.head())

# Display the number of rows in the filtered DataFrame
print(f"\nNumber of rows in the filtered DataFrame: {filtered_df.shape[0]}")
```

```python
# Quantify the number of rows for each value in the 'STATUS_MANIFESTACAO' column
status_counts = filtered_df['STATUS_MANIFESTACAO'].value_counts()

# Display the counts
display(status_counts)
```

```python
# Filter filtered_df for rows where STATUS_MANIFESTACAO is 'EM_ANALISE'
em_analise_df = filtered_df[filtered_df['STATUS_MANIFESTACAO'] == 'EM_ANALISE']

# Quantify the number of rows for each value in the 'NOME_OUVIDORIA_DESTINO' column for the filtered data
nome_ouvidoria_counts_em_analise = em_analise_df['NOME_OUVIDORIA_DESTINO'].value_counts()

# Display the counts
display(nome_ouvidoria_counts_em_analise)
```

```python
# Save the filtered DataFrame to a new CSV file
filtered_df.to_csv('FHEMIG_RH_20250930.csv', index=False)

print("filtered_df saved to filtered_data.csv")
```

```python
# Filter data for 2025
data_2025_df = filtered_df[filtered_df['DATA_MANIFESTACAO'].dt.year == 2025]

# Display the first few rows of the 2025 DataFrame
display(data_2025_df.head())

# Display the number of rows in the 2025 DataFrame
print(f"\nNumber of rows in the 2025 DataFrame: {data_2025_df.shape[0]}")
```

```python
# Calculate top 10 tipificações by ouvidoria for 2025
tipificacao_counts_by_ouvidoria_2025 = data_2025_df.groupby('NOME_OUVIDORIA_DESTINO')['TIPIFICAÇÃO'].value_counts().unstack(fill_value=0)

# Get the top 10 'TIPIFICAÇÃO' values for each 'NOME_OUVIDORIA_DESTINO' for 2024
top_10_tipificacao_by_ouvidoria_2025 = tipificacao_counts_by_ouvidoria_2025.apply(lambda row: row.nlargest(10), axis=1)

# Display the resulting table for 2024
display(top_10_tipificacao_by_ouvidoria_2025)
```

```python
# Transpose the 2024 table
transposed_top_10_2025 = top_10_tipificacao_by_ouvidoria_2025.transpose()

# Calculate the sum of each row (total count for each TIPIFICAÇÃO) for the 2024 data
transposed_top_10_2025['Total'] = transposed_top_10_2025.sum(axis=1)

# Sort the DataFrame by the 'Total' column in descending order
transposed_top_10_sorted_2025 = transposed_top_10_2025.sort_values(by='Total', ascending=False)

# Drop the 'Total' column as it was only used for sorting
transposed_top_10_sorted_2025 = transposed_top_10_sorted_2025.drop(columns='Total')

# Display the transposed and sorted table for 2025
display(transposed_top_10_sorted_2025)
```

```python
# Convert the data types to integer, coercing errors to NaN and filling NaN with 0 for the 2025 data
transposed_top_10_sorted_2025_int = transposed_top_10_sorted_2025.astype(int, errors='ignore').fillna(0).astype(int)

# Save the 2025 DataFrame to an Excel file
transposed_top_10_sorted_2025_int.to_excel('top_tipificacoes_por_ouvidoria_2025.xlsx')

print("Tabela top_tipificacoes_por_ouvidoria_2025.xlsx criada com formato inteiro.")
```

## Filter data for 2024

### Subtask:
Create a DataFrame containing only data from 2024 using the 'DATA_MANIFESTACAO' column.


**Reasoning**:
Create a DataFrame containing only data from 2024 using the 'DATA_MANIFESTACAO' column, and display the first few rows and the count of rows.



```python
# Filter filtered_df for rows where DATA_MANIFESTACAO is in 2024
data_2024_df = filtered_df[filtered_df['DATA_MANIFESTACAO'].dt.year == 2024]

# Display the first few rows of the 2024 DataFrame
display(data_2024_df.head())

# Display the number of rows in the 2024 DataFrame
print(f"\nNumber of rows in the 2024 DataFrame: {data_2024_df.shape[0]}")
```

## Calculate top 10 tipificações by ouvidoria for 2024

### Subtask:
Group the 2024 data by 'NOME_OUVIDORIA_DESTINO' and count the 'TIPIFICAÇÃO' values, then select the top 10 for each ouvidoria.


**Reasoning**:
Group the 2024 data by 'NOME_OUVIDORIA_DESTINO' and count the 'TIPIFICAÇÃO' values, then select the top 10 for each ouvidoria, and store the results in a new DataFrame.



```python
# Group by 'NOME_OUVIDORIA_DESTINO' and count the occurrences of each 'TIPIFICAÇÃO'
tipificacao_counts_by_ouvidoria_2024 = data_2024_df.groupby('NOME_OUVIDORIA_DESTINO')['TIPIFICAÇÃO'].value_counts().unstack(fill_value=0)

# Define a function to get the top N tipificacoes
def get_top_n(row, n=10):
    return row.sort_values(ascending=False).head(n)

# Apply the function to each row to get the top 10 tipificacoes for each ouvidoria
top_10_tipificacao_by_ouvidoria_2024 = tipificacao_counts_by_ouvidoria_2024.apply(get_top_n, axis=1)

# Transpose the table to have 'TIPIFICAÇÃO' in rows and 'NOME_OUVIDORIA_DESTINO' in columns
top_10_tipificacao_by_ouvidoria_2024 = top_10_tipificacao_by_ouvidoria_2024.transpose()

# Display the resulting DataFrame
display(top_10_tipificacao_by_ouvidoria_2024)
```

## Transpose and sort the 2024 table

### Subtask:
Transpose the resulting table and sort the rows by the total count of 'TIPIFICAÇÃO' in descending order.


**Reasoning**:
Calculate the total count for each 'TIPIFICAÇÃO' across all 'NOME_OUVIDORIA_DESTINO' for the 2024 data, sort by this total, and then remove the temporary 'Total' column.



```python
# Calculate the sum of each row (total count for each TIPIFICAÇÃO) for the 2024 data
top_10_tipificacao_by_ouvidoria_2024['Total'] = top_10_tipificacao_by_ouvidoria_2024.sum(axis=1)

# Sort the DataFrame by the 'Total' column in descending order
transposed_top_10_sorted_2024 = top_10_tipificacao_by_ouvidoria_2024.sort_values(by='Total', ascending=False)

# Drop the 'Total' column as it was only used for sorting
transposed_top_10_sorted_2024 = transposed_top_10_sorted_2024.drop(columns='Total')

# Display the transposed and sorted table for 2024
display(transposed_top_10_sorted_2024)
```
