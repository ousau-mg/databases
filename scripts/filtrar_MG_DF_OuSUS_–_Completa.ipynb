```python
import pandas as pd

# Carregar os dados
df = pd.read_csv('/content/OuSUS - Completa.csv')

# Filtrar os dados
df_filtrado = df[df['UF DESTINO ATUAL'].isin(['MG', 'DF'])]

display(df_filtrado.head())
```

```python
# Sobrescrever o arquivo original com os dados filtrados
df_filtrado.to_csv('/content/OuSUS - Completa.csv', index=False)

print("Arquivo original sobrescrito com sucesso, contendo apenas as linhas filtradas.")
```
