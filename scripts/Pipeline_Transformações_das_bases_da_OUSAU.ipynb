## Matriz de Competências/Matriz Saúde
* Necessita das seguintes bases:
  * Siglas
  * Matriz Geográfica
* Utilizada para:
  * OuSUS - Completa
  * SUS Tramitações

```python
# Carregando Dados
Matriz_de_Competências = pd.read_excel("/content/drive/MyDrive/Governo de Minas/OGE/Scripts/Saúde/Power BI/Matriz Saúde.xlsx")

print("Dados carregados com sucesso!")
print(f"Tabela com {Matriz_de_Competências.shape[0]} linhas e {Matriz_de_Competências.shape[1]} colunas")
print("\nPrimeiras linhas:")
Matriz_de_Competências.head()
```

```python
Matriz_de_Competências.columns
```

```python
Matriz_de_Competências.info()
```

```python
# Transformações
MC_modificado = Matriz_de_Competências.copy()

# Ajuste em colunas
MC_modificado.rename(columns={'Destino': 'DESTINO ATUAL'}, inplace=True)
MC_modificado['COMPETÊNCIA 2'] = MC_modificado['Competência'].str.upper()
MC_modificado['MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO'] = MC_modificado['MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO'].str.upper()

# LeftJoin com Siglas
MC_modificado = MC_modificado.merge(Siglas_modificado,
                                    how='left',
                                    left_on=["COMPETÊNCIA 2", "DESTINO ATUAL"],
                                    right_on=["COMPETÊNCIA", "OUVIDORIA OU PONTO DE RESPOSTA"])

MC_modificado.drop(columns=['COMPETÊNCIA', 'OUVIDORIA OU PONTO DE RESPOSTA'], inplace=True)

# Criando coluna de "Filtro - Unidade"
condicoes = [
    MC_modificado['ESFERA'] == "OUTROS",
    MC_modificado['ESFERA'] == "MUNICIPAL",
    MC_modificado['COMPETÊNCIA 2'] == "FHEMIG",
    MC_modificado['COMPETÊNCIA 2'] == "SES REGIONAL"
]

valores = [
    MC_modificado['UF DA OUVIDORIA DE ACOMPANHAMENTO'],
    MC_modificado['MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO'],
    MC_modificado['SIGLA'],
    MC_modificado['MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO']
]

MC_modificado['Filtro - Unidade'] = np.select(condicoes, valores, default=['DESTINO ATUAL'])

# Coluna Mesclada para remover duplicidades
MC_modificado['Mesclado'] = MC_modificado['DESTINO ATUAL'].astype(str) + MC_modificado['MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO'].astype(str)

MC_modificado = MC_modificado.drop_duplicates(subset=['Mesclado'], keep='first')

MC_modificado = MC_modificado.drop(columns=['Mesclado'])

# LeftJoin com Matriz Geográfica
MC_modificado = MC_modificado.merge(MG_modificado,
                                    how='left',
                                    left_on=["MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO", "UF DA OUVIDORIA DE ACOMPANHAMENTO"],
                                    right_on=["MUNICIPIO CIDADAO", "UF CIDADAO"])

MC_modificado['URS'] = MC_modificado['URS'].str.upper()
MC_modificado.rename(columns={'URS': 'GRS'}, inplace=True)

MC_modificado.drop(columns=['Município', 'Micro', 'Macro', 'GEOCODIGO', 'CODMACRO',
                            'MUNICIPIO CIDADAO', 'UF CIDADAO', 'Geocode6', 'Região Intermediária',
                            'CODMICRO', 'Município - MAI', 'POPULAÇÃO'], inplace=True)

# Criando coluna SEOS
condicoes = [
    MC_modificado['Competência'] == "OUTROS",
    MC_modificado['Competência'] == "FEDERAL",
    MC_modificado['Competência'].isna()
]

valores = [
    "OUTROS",
    "OUTROS",
    "OUTROS"
]

MC_modificado['SEOS'] = np.select(condicoes, valores, default=['SEOS'])

MC_modificado = MC_modificado.drop_duplicates(subset=['DESTINO ATUAL'], keep='first').copy()

# Criando coluna Teste
MC_modificado['Teste'] = np.where(
    (MC_modificado['SISTEMA'] == "OUVIDORSUS 2") &
    (MC_modificado['OUVIDORIA VINCULADA - OUVIDORSUS 2'].isna() |
     (MC_modificado['OUVIDORIA VINCULADA - OUVIDORSUS 2'] == "")),
    MC_modificado['DESTINO ATUAL'],
    MC_modificado['OUVIDORIA VINCULADA - OUVIDORSUS 2']
)

MC_modificado.head()
```

```python
# Transformações
SUS_02_modificado = SUS_02.copy()

# Renomeando colunas
SUS_02_modificado.rename(columns={'NUMERO_MANIFESTAÇÃO': 'PROTOCOLO',
                                  'DATA_MANIFESTACAO': 'DATA DA DEMANDA',
                                  'CANAL_DE_ENTRADA': 'MEIO DE ATENDIMENTO',
                                  'ORIGEM_ATENDIMENTO': 'ORIGEM DO ATENDIMENTO',
                                  'STATUS_MANIFESTACAO': 'STATUS',
                                  'CLASSIFICAÇÃO_MANIFESTAÇÃO': 'CLASSIFICACAO',
                                  'NOME_OUVIDORIA_DESTINO': 'DESTINO ATUAL',
                                  'MUNICÍPIO_DO_MANIFESTANTE': 'MUNICIPIO',
                                  'UF_OUVIDORIA_DESTINO': 'UF DESTINO ATUAL',
                                  'ESFERA_DESTINO': 'ESFERA DESTINO ATUAL',
                                  'NIVEL_OUVIDORIA_DESTINO': 'NIVEL DESTINO ATUAL',
                                  }, inplace=True)

# Mudança de tipos
SUS_02_modificado = SUS_02_modificado.astype({
    "NUMERO_ATENDIMENTO": "Int64",
    "CNS_FATO": "Int64",
    "CNPJ_FATO": "Int64"
})

SUS_02_modificado = SUS_02_modificado.astype({
    "NUMERO_ATENDIMENTO": "string",
    "CNS_FATO": "string",
    "CNPJ_FATO": "string"
})

# Adicionando coluna: FECHAMENTO 01
SUS_02_modificado['FECHAMENTO 01'] = SUS_02_modificado['DATA DA DEMANDA'] + pd.to_timedelta(SUS_02_modificado['TEMPO_DE_VIDA'], unit='D')

# Substituindo Valores
substituicoes = {
    "FECHADA": "FECHADO",
    "ARQUIVADA": "ARQUIVADO",
    "EM_ANALISE": "EM ANALISE"
}

SUS_02_modificado["STATUS"] = SUS_02_modificado["STATUS"].replace(substituicoes)

# Adicionando colunas: DATA DO FECHAMENTO, Status - bin, IDENTIFICAÇÃO e SISTEMA
SUS_02_modificado['DATA DO FECHAMENTO'] = np.where(
    SUS_02_modificado['STATUS'].isin(['FECHADO', 'ARQUIVADO']),
    SUS_02_modificado['FECHAMENTO 01'],
    pd.NaT
)

SUS_02_modificado['Status - bin'] = np.where(
    SUS_02_modificado['STATUS'].isin(['FECHADO', 'ARQUIVADO']),
    'ENCERRADA',
    'EM ANDAMENTO'
)

SUS_02_modificado['IDENTIFICAÇÃO'] = np.where(
    SUS_02_modificado['CLASSIFICACAO'].isin(['COMUNICAÇÃO DE IRREGULARIDADE']),
    'ANÔNIMO',
    'IDENTIFICADO'
)

SUS_02_modificado["SISTEMA"] = SUS_02_modificado["PROTOCOLO"].apply(lambda x: "OUSUS 02" if pd.notnull(x) else None)

# Alterando tipos
SUS_02_modificado = SUS_02_modificado.astype({
    "PROTOCOLO": "string",
    "DATA DA DEMANDA": "datetime64[ns]",
    "MEIO DE ATENDIMENTO": "string",
    "CLASSIFICACAO": "string",
    "NUMERO_ATENDIMENTO": "string",
    "TÉCNICO_ATUAL": "string",
    "ORIGEM DO ATENDIMENTO": "string",
    "OUVIDORIA_ORIGEM": "string",
    "MUNICIPIO_OUVIDORIA_ORIGEM": "string",
    "UF_OUVIDORIA_ORIGEM": "string",
    "ESFERA_OUVIDORIA_ORIGEM": "string",
    "DESTINO ATUAL": "string",
    "MUNICIPIO": "string",
    "UF DESTINO ATUAL": "string",
    "ESFERA DESTINO ATUAL": "string",
    "NIVEL DESTINO ATUAL": "string",
    "REFERIDO": "string",
    "LOCAL_FATO": "string",
    "UF_FATO": "string",
    "MUNICIPIO_FATO": "string",
    "SITUACAO_ACOMPANHAMENTO_DESTINO": "string",
    "STATUS": "string",
    "TIPIFICAÇÃO": "string",
    "MARCADOR_TAG": "string",
    "Status - bin": "string",
    "IDENTIFICAÇÃO": "string",
    "SISTEMA": "string",
    "DATA DO FECHAMENTO": "datetime64[ns]"
})

# Selecionando colunas
SUS_02_modificado = SUS_02_modificado[
    [
        "PROTOCOLO",
        "NUMERO_ATENDIMENTO",
        "DATA DA DEMANDA",
        "MEIO DE ATENDIMENTO",
        "CLASSIFICACAO",
        "TÉCNICO_ATUAL",
        "ORIGEM DO ATENDIMENTO",
        "DESTINO ATUAL",
        "MUNICIPIO",
        "UF DESTINO ATUAL",
        "ESFERA DESTINO ATUAL",
        "NIVEL DESTINO ATUAL",
        "REFERIDO",
        "LOCAL_FATO",
        "UF_FATO",
        "MUNICIPIO_FATO",
        "SITUACAO_ACOMPANHAMENTO_DESTINO",
        "STATUS",
        "TIPIFICAÇÃO",
        "MARCADOR_TAG",
        "DATA DO FECHAMENTO",
        "Status - bin",
        "IDENTIFICAÇÃO",
        "SISTEMA",
        "CNS_FATO",
        "CNPJ_FATO"
    ]
]

# Dropando duplicatas de Protocolo
SUS_02_modificado = SUS_02_modificado.drop_duplicates(subset=['PROTOCOLO'], keep='first')

# LeftJoin com Matriz de Competências
SUS_02_modificado = SUS_02_modificado.merge(MC_modificado,
                                            how='left',
                                            left_on="DESTINO ATUAL",
                                            right_on="DESTINO ATUAL")

SUS_02_modificado = SUS_02_modificado.iloc[:, :27]

# Adicionando Coluna de COMPETÊNCIA - DESTINO ATUAL
SUS_02_modificado['COMPETÊNCIA - DESTINO ATUAL'] = (
    SUS_02_modificado['Competência'].fillna(SUS_02_modificado['ESFERA DESTINO ATUAL']))

# Limpeza Final
SUS_02_modificado.drop(columns=['Competência'], inplace=True)
SUS_02_modificado = SUS_02_modificado.drop_duplicates(subset=['PROTOCOLO'], keep='first')

# Substituindo Valores
substituicoes = {
    "COMUNICAÇÃO DE IRREGULARIDADE": "DENÚNCIA",
    "DISSEMINAÇÃO": "INFORMAÇÃO",
    "PEDIDO DE ACESSO A INFORMAÇÃO": "INFORMAÇÃO"
}

SUS_02_modificado["CLASSIFICACAO"] = SUS_02_modificado["CLASSIFICACAO"].replace(substituicoes)

# LeftJoin com OuvidorSUS 2
SUS_02_modificado = SUS_02_modificado.merge(SUS2_Depara_modificado,
                                            how='left',
                                            left_on="TIPIFICAÇÃO",
                                            right_on="TIPIFICAÇÃO")

# Substituindo nulos em ASSUNTO 1
SUS_02_modificado['ASSUNTO 1'] = SUS_02_modificado['ASSUNTO 1'].fillna("NÃO INFORMADO")

# Ordenando Tabela
SUS_02_modificado = SUS_02_modificado.sort_values(by=['DATA DA DEMANDA'],
                                                  ascending=True)

SUS_02_modificado.sample(5)
```

```python
# Transformações
SUS_Tramitacoes_modificado = SUS_Tramitacoes.copy()

# Alterando tipos
SUS_Tramitacoes_modificado = SUS_Tramitacoes_modificado.astype({
    "PROTOCOLO": "int64",
    "SEQUENCIA DO ACOMPANHAMENTO": "int64",
    "STATUS ACOMPANHAMENTO": "string",
    "DATA DO ACOMPANHAMENTO": "datetime64[ns]",
    "OUVIDORIA": "string",
    "MUNICIPIO OUVIDORIA": "string",
    "UF OUVIDORIA": "string"
})

# Ordenando
SUS_Tramitacoes_modificado = SUS_Tramitacoes_modificado.sort_values(by=['PROTOCOLO', 'SEQUENCIA DO ACOMPANHAMENTO'],
                                                                    ascending=[True, True]).reset_index(drop=True)

# Criando colunas para comparação entre linhas consecutivas
SUS_Tramitacoes_modificado['DATA DO ACOMPANHAMENTO.1'] = SUS_Tramitacoes_modificado['DATA DO ACOMPANHAMENTO'].shift(1)
SUS_Tramitacoes_modificado['OUVIDORIA.1'] = SUS_Tramitacoes_modificado['OUVIDORIA'].shift(1)
SUS_Tramitacoes_modificado['USUARIO RESPONSAVEL.1'] = SUS_Tramitacoes_modificado['USUARIO RESPONSAVEL'].shift(1)

# Criando colunas: DataOrigem, Origem, Usuário Origem, Destino, Usuário Destino, Passou pela OGE
SUS_Tramitacoes_modificado["DataOrigem"] = np.where(
    SUS_Tramitacoes_modificado["STATUS ACOMPANHAMENTO"] == "NOVO",
    SUS_Tramitacoes_modificado["DATA DO ACOMPANHAMENTO"],
    SUS_Tramitacoes_modificado["DATA DO ACOMPANHAMENTO.1"]
)

SUS_Tramitacoes_modificado["Origem"] = np.where(
    SUS_Tramitacoes_modificado["STATUS ACOMPANHAMENTO"] == "NOVO",
    "MANIFESTANTE",
    SUS_Tramitacoes_modificado["OUVIDORIA.1"]
)

SUS_Tramitacoes_modificado["Usuário Origem"] = np.where(
    (SUS_Tramitacoes_modificado["SEQUENCIA DO ACOMPANHAMENTO"] == 1) |
    (SUS_Tramitacoes_modificado["STATUS ACOMPANHAMENTO"] == "ARQUIVADO"),
    "MANIFESTANTE",
    SUS_Tramitacoes_modificado["USUARIO RESPONSAVEL.1"]
)

SUS_Tramitacoes_modificado["Destino"] = np.where(
    SUS_Tramitacoes_modificado["STATUS ACOMPANHAMENTO"] == "ARQUIVADO",
    "MANIFESTANTE",
    SUS_Tramitacoes_modificado["OUVIDORIA"]
)

SUS_Tramitacoes_modificado["Usuário Destino"] = np.where(
    SUS_Tramitacoes_modificado["STATUS ACOMPANHAMENTO"].isin(["ARQUIVADO", "FECHADO"]),
    "MANIFESTANTE",
    SUS_Tramitacoes_modificado["USUARIO RESPONSAVEL"]
)

SUS_Tramitacoes_modificado["Passou pela OGE"] = np.where(
    SUS_Tramitacoes_modificado["Destino"] == "OUVIDORIA GERAL DO ESTADO DE MINAS GERAIS - OGE SAÚDE",
    "SIM",
    "NÃO"
)

# Selecionando colunas
SUS_Tramitacoes_modificado = SUS_Tramitacoes_modificado[
    [
        "PROTOCOLO",
        "SEQUENCIA DO ACOMPANHAMENTO",
        "DataOrigem",
        "Origem",
        "STATUS ACOMPANHAMENTO",
        "DATA DO ACOMPANHAMENTO",
        "Usuário Origem",
        "Destino",
        "Usuário Destino",
        "Passou pela OGE"
    ]
]

# Alteranto tipos
SUS_Tramitacoes_modificado = SUS_Tramitacoes_modificado.astype({
    "DataOrigem": "datetime64[ns]"
})

# Renomeando Colunas
SUS_Tramitacoes_modificado = SUS_Tramitacoes_modificado.rename(columns={'DATA DO ACOMPANHAMENTO': 'DataDestino'})

# LeftJoin com Matriz de Competências
SUS_Tramitacoes_modificado = SUS_Tramitacoes_modificado.merge(MC_modificado[["DESTINO ATUAL", "COMPETÊNCIA 2"]],
                                                              how='left',
                                                              left_on="Destino",
                                                              right_on="DESTINO ATUAL").drop(columns=['DESTINO ATUAL'])

# Filtrando linhas onde PROTOCOLO não é nulo nem vazio
SUS_Tramitacoes_modificado = SUS_Tramitacoes_modificado[SUS_Tramitacoes_modificado["PROTOCOLO"].notna() & (SUS_Tramitacoes_modificado["PROTOCOLO"] != "")]

# Ordenando por DataDestino
SUS_Tramitacoes_modificado = SUS_Tramitacoes_modificado.sort_values(by=['DataDestino'],
                                                                    ascending=True)

# Criando coluna Mesclado para remover duplicadas e alterando o tipo de protocolo
SUS_Tramitacoes_modificado["Mesclado"] = (
    SUS_Tramitacoes_modificado["PROTOCOLO"].astype(str) +
    SUS_Tramitacoes_modificado["SEQUENCIA DO ACOMPANHAMENTO"].astype(str)
).astype("int64")

SUS_Tramitacoes_modificado = SUS_Tramitacoes_modificado.drop_duplicates(subset=["Mesclado"])
SUS_Tramitacoes_modificado = SUS_Tramitacoes_modificado.drop(columns=["Mesclado"])
SUS_Tramitacoes_modificado["PROTOCOLO"] = SUS_Tramitacoes_modificado["PROTOCOLO"].astype("string")

SUS_Tramitacoes_modificado.head()
```
