````python
# Transformações
MC_modificado = Matriz_de_Competências.copy()

MC_modificado.rename(columns={'Cometência': 'DESTINO ATUAL'}, inplace=True)
MC_modificado['COMPETÊNCIA 2'] = MC_modificado['Competência'].str.upper()
MC_modificado['MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO'] = MC_modificado['MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO'].str.upper()

# LeftJoin com Siglas
MC_modificado = MC_modificado.merge(Siglas_modificado,
                                    how='left',
                                    left_on=["COMPETÊNCIA 2", "DESTINO ATUAL"],
                                    right_on=["COMPETÊNCIA", "OUVIDORIA OU PONTO DE RESPOSTA"])

MC_modificado.drop(columns=['COMPETÊNCIA', 'OUVIDORIA OU PONTO DE RESPOSTA'], inplace=True)

# Criando coluna de "Filtro - Unidade"
condicoes = [
    MC_modificado['ESFERA'] == "OUTROS",
    MC_modificado['ESFERA'] == "MUNICIPAL",
    MC_modificado['COMPETÊNCIA 2'] == "FHEMIG",
    MC_modificado['COMPETÊNCIA 2'] == "SES REGIONAL"
]

valores = [
    MC_modificado['UF DA OUVIDORIA DE ACOMPANHAMENTO'],
    MC_modificado['MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO'],
    MC_modificado['SIGLA'],
    MC_modificado['MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO']
]

MC_modificado['Filtro - Unidade'] = np.select(condicoes, valores, default=['DESTINO ATUAL'])

# Coluna Mesclada para remover duplicidades
MC_modificado['Mesclado'] = MC_modificado['DESTINO ATUAL'].astype(str) + MC_modificado['MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO'].astype(str)

MC_modificado = MC_modificado.drop_duplicates(subset=['Mesclado'], keep='first')

MC_modificado = MC_modificado.drop(columns=['Mesclado'])

# LeftJoin com Matriz Geográfica
MC_modificado = MC_modificado.merge(MG_modificado,
                                    how='left',
                                    left_on=["MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO", "UF DA OUVIDORIA DE ACOMPANHAMENTO"],
                                    right_on=["MUNICIPIO CIDADAO", "UF CIDADAO"])

MC_modificado['URS'] = MC_modificado['URS'].str.upper()
MC_modificado.rename(columns={'URS': 'GRS'}, inplace=True)

MC_modificado.drop(columns=['Município ', 'Micro', 'Macro', 'GEOCODIGO', 'CODMACRO',
                            'MUNICIPIO CIDADAO', 'UF CIDADAO', 'Geocode6', 'Região Intermediária',
                            'CODMICRO', 'Município  - MAI', 'POPULAÇÃO'], inplace=True)

# Criando coluna SEOS
condicoes = [
    MC_modificado['Competência'] == "OUTROS",
    MC_modificado['Competência'] == "FEDERAL",
    MC_modificado['Competência'].isna()
]

valores = [
    "OUTROS",
    "OUTROS",
    "OUTROS"
]

MC_modificado['SEOS'] = np.select(condicoes, valores, default=['SEOS'])

MC_modificado = MC_modificado.drop_duplicates(subset=['DESTINO ATUAL'], keep='first').copy()

# Criando coluna Teste
MC_modificado['Teste'] = np.where(
    (MC_modificado['SISTEMA'] == "OUVIDORSUS 2") &
    (MC_modificado['OUVIDORIA VINCULADA - OUVIDORSUS 2'].isna() |
     (MC_modificado['OUVIDORIA VINCULADA - OUVIDORSUS 2'] == "")),
    MC_modificado['DESTINO ATUAL'],
    MC_modificado['OUVIDORIA VINCULADA - OUVIDORSUS 2']
)

MC_modificado.head()
````

````python
# Tranformações
SUS_modificado = SUS.copy()

# Mudanças de tipos
SUS_modificado['DATA DA DEMANDA'] = pd.to_datetime(SUS_modificado['DATA DA DEMANDA'])
SUS_modificado['DATA DO FECHAMENTO'] = pd.to_datetime(SUS_modificado['DATA DO FECHAMENTO'])
SUS_modificado['DIAS DE TRAMITACAO'] = SUS_modificado['DIAS DE TRAMITACAO'].astype('Int64')
SUS_modificado['PROTOCOLO'] = SUS_modificado['PROTOCOLO'].astype(str)

# Criando coluna Status - bin
mapeamento_status = {
    'FECHADO': 'ENCERRADA',
    'ARQUIVADO': 'ENCERRADA'
}


SUS_modificado['Status - bin'] = SUS_modificado['STATUS'].map(mapeamento_status).fillna('EM ANDAMENTO')

# Criando coluna IDENTIFICAÇÃO
condicoes = [
    SUS_modificado['SIGILOSO?'] == 'S',
    SUS_modificado['ANONIMO?'] == 'S'
]

valores = [
    'SOB SIGILO',
    'ANÔNIMO'
]

SUS_modificado['IDENTIFICAÇÃO'] = np.select(condicoes, valores, default='IDENTIFICADO')

# LeftJoin com Matriz Saúde
SUS_modificado = SUS_modificado.merge(MC_modificado,
                                      how='left',
                                      left_on=["DESTINO ATUAL", "MUN DESTINO ATUAL"],
                                      right_on=["DESTINO ATUAL", "MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO"])

SUS_modificado.drop(columns=['Competência', 'MUNICIPIO DA OUVIDORIA DE ACOMPANHAMENTO', 'UF DA OUVIDORIA DE ACOMPANHAMENTO',
                             'ESFERA', 'NÍVEL OUVIDORIA', 'SISTEMA', 'OUVIDORIA VINCULADA - OUVIDORSUS 2', 'SIGLA', 'nome_hospitais', 'Filtro - Unidade', 'GRS', 'SEOS', 'Teste'], inplace=True)

SUS_modificado.rename(columns={'COMPETÊNCIA 2': 'COMPETÊNCIA - DESTINO ATUAL'}, inplace=True)

# Substituições de Valores
substituicoes = {
    "INFORMACAO": "INFORMAÇÃO",
    "SUGESTAO": "SUGESTÃO",
    "SOLICITACAO": "SOLICITAÇÃO",
    "RECLAMACAO": "RECLAMAÇÃO",
    "DENUNCIA": "DENÚNCIA"
}

SUS_modificado["CLASSIFICACAO"] = SUS_modificado["CLASSIFICACAO"].replace(substituicoes)

# Filtrando linhas onde PROTOCOLO não é nulo nem vazio
SUS_modificado = SUS_modificado[SUS_modificado["PROTOCOLO"].notna() & (SUS_modificado["PROTOCOLO"] != "")]

SUS_modificado.drop(columns=['ESFERA DESTINO ATUAL', 'SIGILOSO?', 'ANONIMO?', 'DIAS DE TRAMITACAO',], inplace=True)

# Adicionando coluna de SISTEMA
SUS_modificado["SISTEMA"] = SUS_modificado["PROTOCOLO"].apply(
    lambda x: "OUSUS 01" if pd.notna(x) else None
)

SUS_modificado["PROTOCOLO"] = SUS_modificado["PROTOCOLO"].astype("string")

SUS_modificado.rename(columns={'MUNICIPIO CIDADAO': 'MUNICIPIO'}, inplace=True)

SUS_modificado.head()
````
